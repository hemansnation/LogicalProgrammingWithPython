```python
i = 2
while True:
    if i%3 == 0:
        continue
    print(i)
    i += 2
```

**5. WAP to enter a number from user and check it is a palindrome or not.**

121 <br>

```python
x = int(input("Enter a number : "))
s = 0
temp = x
while x > 0:
    r = x%10
    s = s*10 + r
    x = x//10
if s == temp:
    print("Palindrome")
else:
    print("Not Palindrome")
```

**8. WAP to enter a number from user and print individual digits in character formate.**

122 - "One Two Two"

```python
x = int(input("Enter a number : "))
s = 0
while x > 0:
    r = x%10
    s = s*10 + r
    x = x//10

while s > 0:
    r = s%10
    if r == 0:
        print("Zero")
    elif r ==1:
        print("One")
    elif r ==2:
        print("Two")
    elif r ==3:
        print("Three")
    elif r ==4:
        print("Four")
    elif r ==5:
        print("Five")
    elif r ==6:
        print("Six")
    elif r ==7:
        print("Seven")
    elif r ==8:
        print("Eight")
    elif r ==9:
        print("Nine")
    s = s//10
```

**9. WAP for factorial of a number.**

```python
n = int(input("Enter a number : "))
f = 1
while n>1:
    f = f * n    #f = 1, 5, 20, 60, 120
    n -= 1       #n = 5, 4, 3, 2, 1
print("Factorial = ", f)
```

# Loop Else

```python
i = 0
while i < 5:
    print(i)
    i+=1
    if i==3:
        break
else:
    print(0)
```

```python
i = 0
while i < 3:
    print(i)
    i+=1
else:
    print(0)
```

```python
i = 0
while True:
    i += 1
    if i==2:
        continue
    if i==5:
        break
    print(i)
else:
    print("Finished")
```
